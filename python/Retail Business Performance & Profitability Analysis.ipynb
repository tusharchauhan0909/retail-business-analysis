```python
import boto3
```

```python
s3 = boto3.client('s3')

bucket_name = "retail-business-tushar"

files = [
    "fact_orders.csv",
    "dim_products.csv",
    "dim_suppliers.csv",
    "dim_customers.csv"
]

for file in files:
    s3.upload_file(
        Filename=file,
        Bucket=bucket_name,
        Key=f"cleaned/{file}"
    )

print("Clean data uploaded to S3")
```
